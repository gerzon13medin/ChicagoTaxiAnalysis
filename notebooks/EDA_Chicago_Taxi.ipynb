# Hypothesis testing

```python
trips = pd.read_csv('/datasets/project_sql_result_07.csv')
trips.head()
```

```python
trips.describe()
```

```python
# Outliers

# Calculate quartiles and interquartile range (IQR)
Q1 = trips['duration_seconds'].quantile(0.25)
Q3 = trips['duration_seconds'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers
outliers = trips[(trips['duration_seconds'] < lower_bound) | (trips['duration_seconds'] > upper_bound)]
```

```python
outliers
```

```python
plt.figure(figsize=(10, 2))
sns.boxplot(x=trips['duration_seconds'])

# Title and labels
plt.title('Outliers of trip duration', fontsize=14)
plt.xlabel('Duration (seconds)', fontsize=12)

# Show graph
plt.show()
```

```python
trips.shape
```

```python
# Deleting outliers
trips = trips[~trips['duration_seconds'].isin(outliers['duration_seconds'])]
```

```python
trips.shape
```

```python
trips.info()
```

```python
# Become the start_ts to datimetime 
trips['start_ts'] = pd.to_datetime(trips['start_ts'])

# Extract day of the week
trips['day_week'] = trips['start_ts'].dt.day_name()
```

```python
trips['weather_conditions'].unique()
```

```python
# Filtering by bad weather

bad_weather = trips[trips['weather_conditions'] == 'Bad']
bad_weather = bad_weather['duration_seconds']
```

```python
# Filtering by good weather
good_weather = trips[trips['weather_conditions'] == 'Good']
good_weather = good_weather['duration_seconds']
```

Test the hypothesis:

“The average trip duration from Loop to O’Hare International Airport changes on rainy Saturdays.”

H0 - The average trip duration from Loop to O’Hare International Airport is the same 

H1 - The average trip duration from Loop to O’Hare International Airport is different

```python
# We apply Levene's test to determine whether the variances of both populations are equal or different

# H0 - Variance between populations are equeal
# H1 - Variance between populations are different

alpha = 0.05  # Significance level
p_value_levene = stats.levene(bad_weather, good_weather).pvalue

print(p_value_levene)

if p_value_levene < alpha:
    print("We reject the null hypothesis")
else:
    print("We accept the null hypothesis")
```

```python
# Perform an independent two-sample t-test
results = stats.ttest_ind(bad_weather, good_weather, equal_var=True)

# Print the p-value
print('P-value:', results.pvalue)

# Compare the p-value with the significance level
if results.pvalue < alpha:
    print("We reject the null hypothesis")
else:
    print("We accept the null hypothesis")
```

### Hypothesis conclution

On this dataset we detected that there were outliers so, the dataset was cleaned using the interquartile range to get those values and then delete them.

When performing Levene’s test, we found that the variances were equal. This result provided a better reference for subsequently applying the two-sample t-test and setting the equal_var parameter to True. After conducting this hypothesis test, we obtained a result that led us to reject the null hypothesis, indicating that the trip duration from Loop to O’Hare International Airport changes on rainy Saturdays is different
